Print out all (or all varying) hyperparameters of runs in the "results" folder

```python
import os, sys
sys.path.append('..')
import eval_reallabor_utils
import glob

for main_dir in sorted(glob.glob('../results/*')):
    if os.path.isdir(main_dir):
        print(os.path.split(main_dir)[1])
        try:
            hypers = eval_reallabor_utils.get_hypers(main_dir, only_varying=True)
        except FileNotFoundError:
            print('  No model files')
            continue
        for h in hypers:
            print('  '+h.ljust(20)+' '+str(hypers[h]))
```
