```python
import matplotlib.pyplot as plt

architecture = 'shallow'
dim_z = 6
dim_y = 15
dim_x_proj = 6
dim_x = 15
dim_input = 8
n_bases = 0

params = {}
params['A'] = dim_z
if architecture=='shallow':
    params['W'] = dim_z * dim_y * 2
    params['b'] = dim_z + dim_y
else:
    params['W'] = dim_z**2 - dim_z
    params['b'] = dim_z
params['alphas'] = n_bases
params['thetas'] = n_bases * dim_z
params['B'] = dim_x_proj * dim_x
params['C'] = dim_input * dim_z
params['total'] = sum(params.values())

plt.bar(range(len(params)), params.values())
plt.xticks(range(len(params)), params.keys())

params['total']
```

Hierarchized PLRNN

```python
import matplotlib.pyplot as plt

architecture = 'shallow'
dim_z = 12
dim_y = 15
dim_x_proj = 12
dim_x = 15
dim_input = 8
dim_p = 12
n_subjects = 60

params = {}
params['individual'] = dim_p * n_subjects
params['A'] = dim_z * dim_p
if architecture=='shallow':
    params['W'] = dim_z * dim_y * 2 * dim_p
    params['b'] = (dim_z + dim_y) * dim_p
else:
    params['W'] = (dim_z**2 - dim_z) * dim_p
    params['b'] = dim_z * dim_p
params['B'] = dim_x_proj * dim_x * dim_p
params['C'] = dim_input * dim_z * dim_p
params = {p: params[p]//n_subjects for p in params}
params['total'] = sum(params.values())

plt.bar(range(len(params)), params.values())
plt.xticks(range(len(params)), params.keys())

params['total']
```
