```python
import sys
sys.path.append('..')

import glob
import shutil
import os
import utils
from tqdm import tqdm

copy_to = 'eval_dirs'

os.chdir(utils.join_base_path('bptt'))
eval_dirs = glob.glob('results/*/00_summary*')
pbar = tqdm(eval_dirs)
for tree in pbar:
    pbar.set_postfix_str(tree.removeprefix('results/'))
    shutil.copytree(tree, utils.join_base_path(f'bptt/{copy_to}/{tree}'))
```
