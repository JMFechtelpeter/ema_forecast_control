```python
import glob

kalman_folders = glob.glob('results/*Kalman*/*/*/model.pt', recursive=False)

len(kalman_folders)
```

```python
import torch as tc
from tqdm import tqdm
import pickle as pkl

swapped = 0

for i in tqdm(range(len(kalman_folders))):
    swap = False
    params = tc.load(kalman_folders[i])
    with open(kalman_folders[i].replace('model.pt', 'hypers.pkl'), 'rb') as f:
        hypers = pkl.load(f)
    if params['C'].shape[0] == hypers['dim_x'] and params['B'].shape[1] == hypers['dim_s']:
        swap = True
        if params['C'].shape[1] == params['C'].shape[0]:
            if input(f'C quadratic for model {kalman_folders[i]}, swap C and B? (y/n) ') != 'y':
                swap = False
    if swap:
        params['C'], params['B'] = params['B'], params['C']
        tc.save(params, kalman_folders[i])
        swapped += 1

swapped
```
