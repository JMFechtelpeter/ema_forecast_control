```python
import pandas as pd
from collections import defaultdict
import eval_reallabor_utils
import data_utils
import utils

MRT = 2

model_dir = data_utils.join_ordinal_bptt_path(f'results/v2_MRT{MRT}_Transformer_every_day')
model_dirs = eval_reallabor_utils.get_model_folders(model_dir)
tts = pd.read_csv(data_utils.train_test_split_path(MRT, 'first_alarms_no_con_smoothed.csv'), index_col=0)

dates = []
participants = []
wrong_dates = defaultdict(list)
delete_dirs = []
missing_dates = tts.copy()
```

```python
import numpy as np

for dir in model_dirs:

    args = utils.load_args(dir)
    p = args['participant']
    date = float(args['train_on_data_until_timestep'])
    if (missing_dates[p]==date).any():
        missing_dates.loc[missing_dates[p]==date, p] = np.nan
    if not (tts[p]==date).any():
        wrong_dates[p].append(date)
        delete_dirs.append(dir)
```
